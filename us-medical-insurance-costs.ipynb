# U.S. Medical Insurance Costs

This project is based off of a set of US medical insurance company data, stored in the handily named insurance.csv file. This file contains the records of 1339 patients, and has information such as their age, sex, bmi, if they are a smoker, region and finally the cost of their insurance. This project will be geared at determining average insurance costs for each age range, and watching how the different factors affect the cost. 

```python
#importing all necessary modules for this project
import csv
from pandas import * 
from matplotlib import pyplot as plt
```

```python
#creating all variables for the data
ages = []
sexes = []
bmis = []
num_children = []
smoker_statuses = []
regions = []
insurance_charges = []   
```

```python
#reading the csv file
data = read_csv("insurance.csv")
ages = data["age"].tolist()
sexes = data["sex"].tolist()
bmi = data["bmi"].tolist()
num_children = data["children"].tolist()
smoker_statuses = data["smoker"].tolist()
regions = data["region"].tolist()
insurance_charges = data["charges"].tolist()
```

```python
def findAvgChargeAge(age):
    total = 0
    count = 0
    for i in zip(ages, insurance_charges):
        if i[0] == age:
            count += 1
            total += i[1]
    if count == 0:
        return 0
    average = round(total / count, 2)   
    return average
    
```

```python
#create age data points
yCoords = []
xCoords = []
for i in range(min(ages), max(ages)):
    yCoords.append(findAvgChargeAge(i))
    xCoords.append(i)
```

```python
#make graph
plt.plot(xCoords, yCoords)
plt.show()
```

This data shows that the age of a person has a large bearing on the cost for their healthcare, most likely because they are seen as being more likely to have ailments that will cause them to require medical help.
